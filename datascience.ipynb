```python
zip_file_name = "archive.zip"

with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall("data_folder")

print("Files inside zip:")
os.listdir("data_folder")
```

```python
df = pd.read_csv("data_folder/retail_sales_dataset.csv")
df.head()
```

```python
df.info()
df.describe()
df.columns
```

```python
df = df.dropna()
df = df.drop_duplicates()
```

```python
group_data = df.groupby("Product Category")["Total Amount"].sum()
print(group_data)
```

```python
group_data.plot(kind='bar')
plt.title("Purchase by Category")
plt.xlabel("Category")
plt.ylabel("Total Purchase")
plt.show()
```

```python
group_data.plot(kind='pie', autopct='%1.1f%%')
plt.title("Purchase Distribution")
plt.ylabel("")
plt.show()
```
